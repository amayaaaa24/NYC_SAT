# Analyzing Factors Influencing SAT Scores in NYC High Schools
The efficacy of standardized tests and their fairness to certain groups is a highly debated topic in the U.S. educational system. Considering our background knowledge, it would be intriguing to explore the correlations between SAT scores and various demographics. Factors such as race, gender, income, and more can be correlated with SAT scores to gain insights.


The New York City Department of Education provides a wealth of data on high school SAT scores and related demographics. Through their online platform, we have access to datasets on SAT scores by school, school attendance, class size, AP test results, graduation outcomes, demographics, and school surveys. These datasets are interconnected, and in order to explore correlations, we will merge them into a comprehensive dataset. By analyzing these combined datasets, we can gain insights into the factors that influence SAT scores in NYC high schools and better understand the relationships between test performance and various demographic variables.


## Data

- [SAT scores by school](https://data.cityofnewyork.us/Education/2012-SAT-Results/f9bf-2cp4): This dataset provides SAT scores for each high school in New York City.
- [School attendance](https://data.cityofnewyork.us/Education/2010-2011-School-Attendance-and-Enrollment-Statist/7z8d-msnt): It contains attendance information for each school in the city.
- [Class size](https://data.cityofnewyork.us/Education/2010-2011-Class-Size-School-level-detail/urz7-pzb3): This dataset provides information on class sizes for each school.
- [P test results](https://data.cityofnewyork.us/Education/2010-AP-College-Board-School-Level-Results/itfs-ms3e): It includes the results of Advanced Placement (AP) exams taken by students in each high school. Passing an AP exam can earn students college credit in a specific subject.
- [Graduation outcomes](https://data.cityofnewyork.us/Education/2005-2010-Graduation-Outcomes-School-Level/vh2h-md7a): This dataset contains information on the percentage of students who graduated from each school, along with other outcome measures.
- [Demographics](https://data.cityofnewyork.us/Education/2006-2012-School-Demographics-and-Accountability-S/ihfw-zy9j): It provides demographic information for each school, such as student ethnicity and socio-economic background.
- [School survey](https://data.cityofnewyork.us/Education/2011-NYC-School-Survey/mnz3-dyi8): Surveys conducted with parents, teachers, and students at each school to gather their feedback and perceptions.
https://data.cityofnewyork.us/Education/2011-NYC-School-Survey/mnz3-dyi8


### [1] Read in Data

```python
# Read in data
import pandas as pd
import numpy
import re

data_files = [
    "ap_2010.csv",
    "class_size.csv",
    "demographics.csv",
    "graduation.csv",
    "hs_directory.csv",
    "sat_results.csv"
]

data = {}

for f in data_files:
    d = pd.read_csv("schools/{0}".format(f))
    data[f.replace(".csv", "")] = d
```

### [2] Read in the surveys

```python
all_survey = pd.read_csv("schools/survey_all.txt", delimiter="\t", encoding='windows-1252')
d75_survey = pd.read_csv("schools/survey_d75.txt", delimiter="\t", encoding='windows-1252')
survey = pd.concat([all_survey, d75_survey], axis=0)

survey["DBN"] = survey["dbn"]

survey_fields = [
    "DBN", 
    "rr_s", 
    "rr_t", 
    "rr_p", 
    "N_s", 
    "N_t", 
    "N_p", 
    "saf_p_11", 
    "com_p_11", 
    "eng_p_11", 
    "aca_p_11", 
    "saf_t_11", 
    "com_t_11", 
    "eng_t_11", 
    "aca_t_11", 
    "saf_s_11", 
    "com_s_11", 
    "eng_s_11", 
    "aca_s_11", 
    "saf_tot_11", 
    "com_tot_11", 
    "eng_tot_11", 
    "aca_tot_11",
]
survey = survey.loc[:,survey_fields]
data["survey"] = survey
```

## Data Cleaning
### [1] Add DBN columns

```python
data["hs_directory"]["DBN"] = data["hs_directory"]["dbn"]

def pad_csd(num):
    string_representation = str(num)
    if len(string_representation) > 1:
        return string_representation
    else:
        return "0" + string_representation
    
data["class_size"]["padded_csd"] = data["class_size"]["CSD"].apply(pad_csd)
data["class_size"]["DBN"] = data["class_size"]["padded_csd"] + data["class_size"]["SCHOOL CODE"]
```

### [2] Convert columns to numeric

```python
cols = ['SAT Math Avg. Score', 'SAT Critical Reading Avg. Score', 'SAT Writing Avg. Score']
for c in cols:
    data["sat_results"][c] = pd.to_numeric(data["sat_results"][c], errors="coerce")

data['sat_results']['sat_score'] = data['sat_results'][cols[0]] + data['sat_results'][cols[1]] + data['sat_results'][cols[2]]

def find_lat(loc):
    coords = re.findall("\(.+, .+\)", loc)
    lat = coords[0].split(",")[0].replace("(", "")
    return lat

def find_lon(loc):
    coords = re.findall("\(.+, .+\)", loc)
    lon = coords[0].split(",")[1].replace(")", "").strip()
    return lon

data["hs_directory"]["lat"] = data["hs_directory"]["Location 1"].apply(find_lat)
data["hs_directory"]["lon"] = data["hs_directory"]["Location 1"].apply(find_lon)

data["hs_directory"]["lat"] = pd.to_numeric(data["hs_directory"]["lat"], errors="coerce")
data["hs_directory"]["lon"] = pd.to_numeric(data["hs_directory"]["lon"], errors="coerce")
```

### [3] Condense datasets

```python
class_size = data["class_size"]
class_size = class_size[class_size["GRADE "] == "09-12"]
class_size = class_size[class_size["PROGRAM TYPE"] == "GEN ED"]

class_size = class_size.groupby("DBN").agg(numpy.mean)
class_size.reset_index(inplace=True)
data["class_size"] = class_size

data["demographics"] = data["demographics"][data["demographics"]["schoolyear"] == 20112012]

data["graduation"] = data["graduation"][data["graduation"]["Cohort"] == "2006"]
data["graduation"] = data["graduation"][data["graduation"]["Demographic"] == "Total Cohort"]
```

### [4] Convert AP scores to numeric

```python
cols = ['AP Test Takers ', 'Total Exams Taken', 'Number of Exams with scores 3 4 or 5']

for col in cols:
    data["ap_2010"][col] = pd.to_numeric(data["ap_2010"][col], errors="coerce")
```

### [5] Combine the datasets

```python
combined = data["sat_results"]

combined = combined.merge(data["ap_2010"], on="DBN", how="left")
combined = combined.merge(data["graduation"], on="DBN", how="left")

to_merge = ["class_size", "demographics", "survey", "hs_directory"]

for m in to_merge:
    combined = combined.merge(data[m], on="DBN", how="inner")

combined = combined.fillna(combined.mean())
combined = combined.fillna(0)
```

### [6] Add a school district column for mapping

```python
def get_first_two_chars(dbn):
    return dbn[0:2]

combined["school_dist"] = combined["DBN"].apply(get_first_two_chars)
```

## Data Analysis
### [1] Find correlations

```python
correlations = combined.corr()
correlations = correlations["sat_score"]
print(correlations)
```

### [2] Plotting survey correlations

```python
# Remove DBN since it's a unique identifier, not a useful numerical value for correlation.
survey_fields.remove("DBN")
```

```python
# Set up matplotlib to work in Jupyter notebook.
%matplotlib inline
import matplotlib.pyplot as plt
```

Name of the survey fields and their corresponding explaination:




```python
combined[survey_fields].corrwith(combined['sat_score']).sort_values().plot.barh(figsize=(10,9))
plt.xlabel('Correlation Coefficient')
plt.ylabel('Survey')
plt.show()
```

The highest correlations are the number of the parent respondents and the student respondents, which are followed by academic expectation score based on student responses, safety and respect total score, safety and respect score based on student responses, and safety and respect score based on teacher responses.

It's clear that whether the school offers a environment with safety and respect has a high correlation with the SAT score.

Most of the data points concentrate around SAT score eqaul to 1200 but appears to have a slightly increasing trend.

Now, let's determine whether there are any racial differences in SAT performance. The columns that indicate the percentage of each race at a given school includes:
- white_per
- asian_per
- black_per
- hispanic_per

We will examine the correlation of these columns with `sat_score`.

```python
# a bar plot of the correlations between the columns above and sat_score
combined.loc[:,['white_per','asian_per','black_per','hispanic_per']].corrwith(combined['sat_score']).sort_values().plot.barh(figsize=(5,3))
plt.xlabel('Correlation Coefficient')
plt.ylabel('Races')
plt.show()
```

Schools with higher percentage of white and asian students appear to have higher SAT score while schools with more black and Hispanic students have lower SAT scores.

```python
combined.plot.scatter(x='hispanic_per',y='sat_score')
plt.show()
```

Schools with a higher percentage of Hispanic students tend to have lower SAT scores, although the majority of the data is concentrated around 1200.

```python
# School with a hispanic_per greater than 95%
combined[combined['hispanic_per'] > 95]['SCHOOL NAME']
```

Half of the high schools with more than 95 percent Hispanic students are international schools.

```python
# School with a hispanic_per less than 10% and an average SAT score greater than 1800
combined[(combined['hispanic_per'] < 10) & (combined['sat_score'] > 1800)]['SCHOOL NAME']
```

Schools with a Hispanic student percentage of less than 10 percent and an average SAT score of more than 1800 are among the top 10 high schools in New York City, according to the report from [U.S. News](https://www.usnews.com/education/best-high-schools/new-york).

```python
# Gender percentage vs. SAT score
combined.loc[:,['male_per','female_per']].corrwith(combined['sat_score']).sort_values().plot.barh(figsize=(5,3))
plt.xlabel('Correlation Coefficient')
plt.ylabel('Races')
plt.show()
```

The percentage of females has a stronger correlation with SAT scores compared to the percentage of males. This suggests that females may have better performance on the SAT.

```python
# A scatter plot of female_per vs. sat_score
combined.plot.scatter(x='female_per',y='sat_score')
plt.show()
```

High schools with a female percentage between 40 and 80 tend to have higher SAT scores. The majority of the data is concentrated within the range of 40 to 60 percent.

High schools in New York City with female percentage greater than 60 and average SAT score greater than 1700 including:

```python
mask = (combined['female_per'] > 60) & (combined['sat_score'] > 1700)
combined.loc[mask,'SCHOOL NAME']
```

In the United States, high school students have the opportunity to take Advanced Placement (AP) exams, which can earn them college credit in various subjects. It is logical to assume that the number of students who take AP exams at a school would have a strong correlation with the school's SAT scores. To ensure unbiased results, we will focus on the percentage of students at each school who have taken at least one AP exam, rather than relying solely on total enrollment as a factor.

```python
# Calculate the percentage of students in each school that took an AP exam.
ap_per = combined['AP Test Takers ']/combined['total_enrollment']

# Investigate the relationship between AP scores and SAT scores.
plt.scatter(ap_per,combined['sat_score'])
plt.xlabel('AP Test Takers Percentage')
plt.ylabel('SAT Score')
plt.show()
```

The average SAT score tends to increase when the percentage of AP test takers is less than 60 percent. The majority of the data points are concentrated around a score of 1200.
